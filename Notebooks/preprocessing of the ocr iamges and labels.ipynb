```python
#Creer un fichier txt qui liste le full path des images qui serviront au train 
#creer un fichier txt qui liste le full path des images qui serviront au validation 
```

```python
def create_dataset(mypath, myratio): #create a function with mypath (image folder) and myratio as variables
    all_images = glob2.glob(mypath)  #list all images 
    n_max = int(myratio * len(all_images)) #integer corresponsding to the index of the last training element
    train_images = all_images[:n_max]  # from the start index to n_max
    valid_images = all_images[n_max:]  #from n_max to the end
    
    return train_images, valid_images 
```

```python
train_list, valid_list = create_dataset("C:\\Users\\Light\\Desktop\\ocr\\*.png", 0.8)
```

```python
import pandas as pd 
```

```python
train_frame = pd.DataFrame(train_list)
valid_frame = pd.DataFrame(train_list)
```

```python
train_frame.to_csv(r'C:\Users\Light\Desktop\ocr\train.txt', header=None, index=None, sep=' ')
valid_frame.to_csv(r'C:\Users\Light\Desktop\ocr\valid.txt', header=None, index=None, sep=' ')
```
